Importing Dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from xgboost import XGBRegressor
from sklearn import metrics
```

Import Boston House Price data set

```python
 house_price_dataset = pd.read_csv('/content/boston_house_prices.csv')
```

```python
house_price_dataset
```

```python
#check for missing values

house_price_dataset.isnull().sum()
```

```python
# understanding the dataset

house_price_dataset.describe()
```

```python
correalation = house_price_dataset.corr()
```

```python
# constructing heat map to undersatnd correlation

plt.figure(figsize = (10,10))
sns.heatmap(correalation,cbar = True, square = True, fmt = '.1f', annot = True, annot_kws = {'size':8}, cmap = 'Blues')
```

Splitting the dataset


```python
X = house_price_dataset.drop(['MEDV'], axis = 1)
Y = house_price_dataset['MEDV']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

XG Boost Regressor

```python
#load the model

model = XGBRegressor()
```

```python
#training th model with X_train

model.fit(X_train, Y_train)
```

Evaluation

Prediction on Training data

```python
#accuracy on training data

training_data_prediction = model.predict(X_train)
```

```python
training_data_prediction
```

Comparing the values predicted by model on training data with the correct values


So //training_data_prediction// & //Y_train// will be compared


```python
# R squared error

score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error

score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)
```

```python
print('R squared error : ', score_1)
print('Mean Absolute Error : ', score_2)
```

Data Visualisation on Training Data

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

Prediction on Test Data

```python
#accuracy for test data pred

test_data_prediction = model.predict(X_test)
```

```python
# R squared error

score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error

score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)
```

```python
print('R squared error : ', score_1)
print('Mean Absolute Error : ', score_2)
```

Data Visualisation on Trest Data

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
